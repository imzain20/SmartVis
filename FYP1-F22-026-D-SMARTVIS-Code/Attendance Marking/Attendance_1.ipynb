```python

from tkinter import *
import cv2
import os
import numpy as np
import time
import sys
from PIL import Image;
from PIL import ImageTk, Image
from datetime import datetime
from datetime import date
 
# Returns the current local date
today = date.today()
today=str(today)

current_datetime = datetime.now()
current_datetime = str(current_datetime)

file_name = (today+".txt")

ids = input('Enter your Roll no. id: ')


win= Tk()

#for window title
win.title("AUTOMATED ATTENDANCE")


#frame size
win.geometry("1280x1280")
win.configure(bg='purple')
frame = Frame(win, width=20, height=20)
frame.pack()
frame.place(anchor='center', relx=0.5, rely=0.2)

#for bg image
bg = PhotoImage(file = "bg1-overlay.png")
  
# Show image using label
label1 = Label( win, image = bg)
label1.place(x = 0, y = 0)

# Create an object of tkinter ImageTk
img = ImageTk.PhotoImage(Image.open("faceid.JPG"))

# Create a Label Widget to display the text or Image
label = Label(frame, image = img)
label.pack()

Label(win, text="AUTOMATED ATTENDANCE", bg='black', fg='white', font=('Montserrat 20 bold')).pack(pady=4)

#ids=2

'''

#Initialize a Label to display the User Input
label=Label(win, text="")
label.pack()


#Create an Entry widget to accept User Input
entry= Entry(win, width= 24)
entry.focus_set()
entry.pack()
entry.place(x=610,y=220)
'''
#path assurance function
def assure_path_exists(path):
    dir = os.path.dirname(path)
    if not os.path.exists(dir):
        os.makedirs(dir)

def display_text():
    
    global entry
    ids= entry.get()
    label.configure(text=ids)
    ids=int(ids)
    print(ids)



#dataset capture funtion
def faceCapture():
    
    vid_cam = cv2.VideoCapture(0)  #video cam start
    face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')    # Detect face in video by Haarcascade Frontal Face

    count = 0
    assure_path_exists("Dataset_new1/")

    #for capturing face
    while (True):

        # Capture video frame
        _, image_frame = vid_cam.read()

        # Convert into grayscale
        gray = cv2.cvtColor(image_frame, cv2.COLOR_BGR2GRAY)

        # Detect frames of different sizes, list of faces rectangles
        face_img = face_detector.detectMultiScale(gray, 1.3, 5)

        # Loops for each faces
        for (x, y, w, h) in face_img:
            # rectangle crop
            cv2.rectangle(image_frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
            # sample face increment
            count += 1
            # insering images into dataset
            cv2.imwrite("Dataset_new1/Student." + str(ids) + '.' + str(count) + ".jpg", gray[y:y + h, x:x + w])
            # displaying human face
            cv2.imshow('frame', image_frame)
            
            # To stop taking video, press 'q' 
        if cv2.waitKey(100) & 0xFF == ord('q'):
            break
            #set number of frames for pics
        elif count >= 50:
            print("Face Capture Success")
            break
        
            
    # Stop video
    vid_cam.release()
    cv2.destroyAllWindows()
    
    
    #training
    
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml');


    def getImagesAndLabels(path):
        # accessing all files of folder
        imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
        face_Samp = []
        IdsArr = []
        #loading data
        for imagePath in imagePaths:
            # loading images 
            #grayscale conversion
            pilImage = Image.open(imagePath).convert('L')
            # adding it to numpyarray
            imageNp = np.array(pilImage, 'uint8')
            # id getting
            Id = int(os.path.split(imagePath)[-1].split(".")[1])
            # face attraction
            face_img = detector.detectMultiScale(imageNp)
            # appending faces and ids
            for (x, y, w, h) in face_img:
                face_Samp.append(imageNp[y:y + h, x:x + w])
                IdsArr.append(Id)
        
        return face_Samp, IdsArr
    faces,Ids = getImagesAndLabels('Dataset_new1')   #function calling
    s = recognizer.train(faces, np.array(Ids))
    print("Training Success")
    recognizer.write('trainer.yml')

    

#attendance marking function
def attendance():
    start = time.time()
    period = 8
    
    file1 = open(file_name, 'w')
    file1.write("Attendance for " + current_datetime + "\n\n")
    
    period = 8
    face_cas = cv2.CascadeClassifier(
        'haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0);
    recognizer = cv2.face.LBPHFaceRecognizer_create();
    #trainer file
    recognizer.read('trainer.yml');

    id = 0;
    filename = 'filename.txt';
    dict = {
        'item1': 1
    }

    font = cv2.FONT_HERSHEY_SIMPLEX
    while True:
        ret, img = cap.read();
        #images conversion to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY);
        faces = face_cas.detectMultiScale(gray, 1.3, 7);
        for (x, y, w, h) in faces:
            roi_gray = gray[y:y + h, x:x + w]
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2);
            id, conf = recognizer.predict(roi_gray)
            if (conf < 50):
                if (id == 1):
                    id = 'Dexter'
                    if ((str(id)) not in dict):
                        dict[str(id)] = str(id);
                        file1 = open(file_name,"a")
                        file1.write(id + "     Present\n")
                        print(id + "     Present\n")
                         

                elif (id == 2):
                    id = 'Khan'
                    if ((str(id)) not in dict):
                        dict[str(id)] = str(id);
                        file1 = open(file_name,"a")
                        file1.write(id + "     Present\n")
                        print(id + "     Present\n")
                        

                elif (id == 3):
                    id = 'Zain'
                    if ((str(id)) not in dict):
                        dict[str(id)] = str(id);
                        file1 = open(file_name,"a")
                        file1.write(id + "     Present\n")
                        print(id + "     Present\n")
                        
                elif (id == 4):
                    id = 'Danial'
                    if ((str(id)) not in dict):
                        dict[str(id)] = str(id);
                        file1 = open(file_name,"a")
                        file1.write(id + "     Present\n")
                        print(id + "     Present\n")

            else:
                id = 'Unknown'
                file1 = open(file_name,"a")
                file1.write(id + "     Absent\n")
                print(id + "     Absent\n")
                break

       #     cv2.putText(img, str(ids) + " " + str(conf), (x, y - 10), font, 0.55, (120, 255, 120), 1)
       # cv2.imshow('frame', img);

        if time.time() > start + period:
            break;
        if cv2.waitKey(100) & 0xFF == ord('q'):
            break;

    cap.release();
    cv2.destroyAllWindows()

    
def quit():
    #print(ids)
    win.destroy()
    
def callback():
   Label(win, text="Place your face infront of the camera...", font=('arial 12')).pack(pady=3)




#btn0=Button(win, text= "Enter ID",width= 20, command= display_text, bg='black', fg='white')
btn=Button(win, text="Face Capture",width= 20, command= faceCapture, bg='black', fg='white')
#btn1=Button(win, text="Start Storing Identity",width= 20, command= training, bg='black', fg='white')
btn2=Button(win, text="Mark My Attendance",width= 20, command= attendance, bg='black', fg='white')
btn3=Button(win, text="Exit",width= 20, command= quit, bg='black', fg='white')
# Set the position of button to coordinate (100, 20)

#btn0.place(x=610, y=250)
btn.place(x=610, y=300)
#btn1.place(x=610, y=350)
btn2.place(x=610, y=350)
btn3.place(x=610, y=400)
#win.bind('<Return>',lambda event:display_text())
win.bind('<Return>',lambda event:faceCapture())
#win.bind('<Return>',lambda event:training())
win.bind('<Return>',lambda event:attendance())
win.bind('<Return>',lambda event:quit())


win.mainloop()



```
